```python
#create a list of 10 integer elements print their squares save these squares ina new list
l1=[12,22,34,45,56,62,76,88,91,10]
l3=[]
for i in l1:
  l3.append(i**2)
print(l3)

```

```python
#list comprehension
#newlist = [output for]
l1=[12,22,34,45,56,62,76,88,91,10 ]
l3=[i*i for i in l1 ]
print(l3)
```

```python
#checking a condition in a list comprehension
#newlist = [output for if cond]
l1=[12,22,34,45,56,62,76,88,91,10 ]
l3=[i*i for i in l1 if i>50]
print(l3)
```

```python
#LC
#newlist = [(st of if) if (cond) else (st of else) for]
l1=[12,56,22,34,62,76,88,91,42,10 ]
l3=[i*i if i>50 else 'X' for i in l1 ]
print(l3)
```
