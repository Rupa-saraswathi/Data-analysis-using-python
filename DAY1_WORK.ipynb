```python
#creating a tuple
t=(1,"python",2.56,'a',59,20,'data')
print(t)
```

```python
# tuple slicing
print(t[0])
print(t[2:])
print(t[:-1])
print(t[1:3])
print(t[:])
```

```python
#updating a tuple
t1=(1,2,35,4,5,3,5,2,5)
t2=('a','b','c','d')
print(t1+t2)
```

```python
#deleting a tuple
del(t)
```

```python
#operations on a tuple
print(len(t1))   #returns length of tuple
print(max(t1))
print(min(t1))
print(t1*2)
print(t1.count(5))
print(t1.index(35))
```

```python
t3=(92,3,4,5,7,6)
print(5 in t1)    #checks
print(t1>t3)       #comparision
print(tuple("hello"))
print(tuple([23,45,6,7,89,90]))
```

```python
t4=(('hi','python','hello','data'),('c','java','c++'))
for i in t4:
  print(i)
```

```python
#combining tuple and list
t5=(1,2,3,4,5)
l=[6,7,8,9,10]
print(list(zip(t5,l)))
```

```python
#accessing elements in tuple
for i in t3:
  print(i)
```

```python

tup=(9,8,7,6,5,4,3,2,1)
for i in tup:
  if i%2==0:
    print(i**2)
```

```python
#creating a set
s={1,2,3,4}
print(s)
```

```python
#converting into a set
s1=set([1,'apple',6.9,42,"set"])
print(s1)
```

```python
#using for loop
s2={5,3,7,8,4}
for i in s2:
  print(i,end=" ")
```

```python
#using if-else
if 4 in s:
  print(" 4 is present")
else:
  print("4 is not present")
```

```python
#set functions
s3={1,44,56,4,3,77}
s4={10,50,30,30,20,60,6,73,89}
print(s3.union(s4))
print(s3.intersection(s4))
print(s3.difference(s4))
print(s3.symmetric_difference(s4))
```

```python
#checks
print(s3.issubset(s4))
print(s3.issuperset(s4))
```

```python
#set manipulations
print(s1.add(10))
print(s1.update(s2))
print(s3.discard(77))
print(s1.pop())
print(s1)
print(s3)
```

```python
#set  clear

print(s1.clear())
```

```python
#set comprehension
l=[1,2,3,4,5]
s={i**2 for i in l}
print(s)
```

```python
#using if
l1=[10,20,34,50,69]
s1={i*5  if i%2==0 else i*3  for i in l1}
print(s1)
```
