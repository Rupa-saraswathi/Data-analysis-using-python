```python
df = pd.read_csv("/content/cancer patient data sets.csv")
```

```python
feature_cols = ['Dry Cough' ,'Frequent Cold','Obesity','Smoking']
features = df[feature_cols]
Target = df.Level
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(features,Target)
```

```python
df
```

```python
model.predict([[3,2,4,3]])
```

```python
dc = int(input("enter dry cough count: "))
fc = int(input("enter frequently cold count: "))
o = int(input("enter obesity count: "))
s = int(input("enter smoking count: "))
pred = model.predict([[dc,fc,o,s]])
if pred == "Low":
  print("less chance of cancer")
elif pred == "High":
  print("high chances of cancer")
else :
  print("medium chances of cancer")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
from sklearn.ensemble import  RandomForestClassifier
RF = RandomForestClassifier(n_estimators=10)
```

```python
df1 = pd.read_csv("/content/cancer patient data sets.csv")
```

```python
feature_cols = ['Dry Cough' ,'Frequent Cold','Obesity','Smoking']
features = df1[feature_cols]
Target = df1.Level
```

```python
RF.fit(features,Target)
RF.predict([[3,2,4,3]])
```

```python
dc = int(input("enter dry cough count: "))
fc = int(input("enter frequently cold count: "))
o = int(input("enter obesity count: "))
s = int(input("enter smoking count: "))
pred = model.predict([[dc,fc,o,s]])
if pred == "Low":
  print("less chance of cancer")
elif pred == "High":
  print("high chances of cancer")
else :
  print("medium chances of cancer")
```

```python
import os
output_dir = "tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
RF.estimators_
```

```python
for i,tree in enumerate(RF.estimators_):
  tree_dot_file = os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file = os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['Dry Cough' ,'Frequent Cold','Obesity','Smoking'],
                    class_names=[str(cls) for cls in RF.classes_],filled=True,rounded=True)
```

```python
command = f"dot -Tpng {tree_dot_file} -o {tree_png_file} "
os.system(command)
print(f"Tree (i) visualisation saved to {tree_png_file}")
```
