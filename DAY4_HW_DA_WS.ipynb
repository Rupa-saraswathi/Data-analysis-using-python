```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
options = webdriver.ChromeOptions()
```

```python
driver = webdriver.Chrome(options = options)
```

```python
driver.get("https://www.amazon.in/")
```

```python
search = driver.find_element(By.XPATH,".//input[@class='nav-input nav-progressive-attribute']")
```

```python
search.send_keys("top 10 dell laptops")
```

```python
search.send_keys(Keys.ENTER)
```

```python
brand = driver.find_elements(By.XPATH,".//span[@class='a-size-medium a-color-base a-text-normal']")
brands = [ i.text  for i in brand]
```

```python
print(brands)
print(len(brands))
```

```python
price = driver.find_elements(By.XPATH,".//span[@class='a-price-whole']")
prices = [ i.text for i in price]
```

```python
print(prices)
print(len(prices))
```

```python
df = pd.DataFrame(columns = ["brand" , "Price"])
df
```

```python
df['brand'] = brands
df['Price'] = prices
```

```python
df
```

```python
df.to_csv("laptops.csv")
```

```python
watch_name = driver.find_elements(By.XPATH,".//h3[@class='product-brand']")
price = driver.find_elements(By.XPATH,".//span[@class='product-discountedPrice']")
```

```python
brands = [i.text for i in watch_name]
prices = [i.text for i in price]
print(len(brands),len(prices))
```

```python
print(brands)
print(prices)
```

```python
df1 = pd.DataFrame(columns = ['Brands' , 'Price'])
```

```python
df1["Brands"]=brands
df1["Price"]=prices
print(df1)
```
