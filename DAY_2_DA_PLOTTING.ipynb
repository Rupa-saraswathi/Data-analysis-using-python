QUESTION:
there are 5 subjects psychology ,nanotechnology,medicine,oncology,anthropology ,2 students have applied to do masters in any of these 5 fields with a line plot show in which field
they can excel.the cutoff for each subject is 70 s1=[67,45,73,71,57];s2=[43,67,68,53,85]

```python
subjects=['psychology','nano technology','medicine','oncology','anthropology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
plt.title("Student marks")
#plt.plottype(X,Y,label,color,marker,linewidth,linestyle)
#student 1
plt.scatter(subjects,s1,label="Student 1",color ="k",marker='p') #use plot in scatter place
#student 2
plt.scatter(subjects,s2,label="Student 2",color ="b",marker='H')
plt.xlabel("subjects")
plt.ylabel("marks")
#cutoff
plt.plot(subjects,cut,label="cutoff",color="r",linestyle=":",linewidth=4)
plt.legend()
```

QUESTION:
an india vs australia match is on live ,india has given 298 as target for australia that runs for every 5 overs is given as follows:
[25,51,84,131,160,189,220,250,267,297] australia have completed 25 overs
Austarlia = [15,41,94,110,151] plot overs vs runs for both the countries with india as blue line and australia as yellow line with required labels and legend.
the marker for india is flag and the marker australia is a "+".take overs from arange function create arrays from every run ,linewidth=2

```python
import numpy as np
india=[25,51,84,131,160,189,220,250,267,297]
australia= [15,41,94,110,151]
target=[298,298,298,298,298,298,298,298,298,298]
overI=np.arange(5,51,5)
overA=np.arange(5,26,5)
plt.plot(overI,india,label="India",color='b',marker='>',linewidth=2)
plt.plot(overA,australia,label="Australia",color='y',marker='+',linewidth=2)
plt.xlabel("Overs")
plt.ylabel("runs")
plt.title("India vs Australia")
plt.legend()
plt.grid()
```
