seaborn has many of its own csv files like :fmri,tips,dowjones,healthexp,dots
showing the difference using color called hue

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.relplot(data=tips, x= "total_bill",y = "tip" , hue ="smoker")
#hue - creating difference based on a column via colors
```

```python
sns.relplot(data=tips, x= "total_bill",y = "tip" , hue ="sex")
```

```python
sns.relplot(data=tips,x= "total_bill",y = "tip" , hue ="smoker" , style = "time", palette="bright")
```

```python
sns.relplot(data=tips,x= "total_bill",y = "tip" , hue ="size")
```

```python
dj = sns.load_dataset("dowjones")
```

```python
sns.relplot(data=dj, x="Date", y= "Price",kind="line")
# by default relplot have scatter plot if we dont mention kind it gives defaultly scatter
```
